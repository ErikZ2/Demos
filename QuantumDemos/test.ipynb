```python
import numpy as np

import matplotlib.pyplot as plt

# Generate a range of x values
x = np.linspace(0, 2 * np.pi, 100)

# Compute the corresponding sine values
y = np.sin(x)

# Create the plot
plt.plot(x, y)

# Set the labels and title
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Plot of the Sine Function')

# Display the plot
plt.show()
```
